```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([[0, 0],
             [0, 1],
             [1, 0],
             [1, 1]
             ])

y = np.array([[0], [1], [1], [0]])
```

```python
X.shape, y.shape
```

```python
def h1(X):
    z = 1*X[:,0] + 1*X[:,1] + 0
    a = z.copy()
    a[a<0] = 0
    return a

def h2(X):
    z = 1*X[:,0] + 1*X[:,1] -1
    a = z.copy()
    a[a<0] = 0
    return a
```

```python
def h3(X):
    z = 1*h1(X) - 2*h2(X)
    a = z.copy()
    a[a<0] = 0
    return a
    
```
