```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])
```

```python
img = mpimg.imread('iitgn.jpeg')     
gray = rgb2gray(img)   
```

```python
plt.imshow(img)
```

```python
plt.imshow(gray,cmap='Greys')
```

```python
gray.shape
```

```python
gray = gray[:, :400]
```

```python
plt.imshow(gray,cmap='Greys')
```

```python
from keras import Sequential
from keras.layers import Conv2D

model = Sequential()
model.add(Conv2D(1, kernel_size=(3, 3), strides=(1, 1),
                 activation='linear',
                 input_shape=(400,400, 1)))
l = model.layers[0]
vert_filter = np.array(([[1, 0, -1],
                         [1, 0, -1],
                         [1, 0, -1]])).reshape(3, 3, 1, 1)
l.set_weights([vert_filter, np.array([0])])
```

```python
plt.imshow(model.predict(gray.reshape(1,400,400,1)).reshape(398,398),cmap='Greys')
```

```python
l = model.layers[0]
vert_filter = np.array(([[1, 1, 1],
                         [0,0, 0],
                         [-1 , -1, -1]])).reshape(3, 3, 1, 1)
l.set_weights([vert_filter, np.array([0])])
```

```python
plt.imshow(model.predict(gray.reshape(1,400,400,1)).reshape(398,398),cmap='Greys')
```
