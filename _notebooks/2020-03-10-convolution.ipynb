```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib.patches as patches
fig, ax = plt.subplots()
# Create a Rectangle patch


nx, ny = 5, 5
gap = 10
ar = np.round(np.random.randn(nx, ny), 1)
rects = np.zeros((nx, ny)).tolist()
for i in range(nx):
    for j in range(ny):
        rects[i][j] = patches.Rectangle((i*gap, j*gap),gap,gap,linewidth=1,edgecolor='k',facecolor='none')
        ax.add_patch(rects[i][j])
        ax.text(i*gap+gap/3, j*gap+gap/3, ar[i, j])


# Add the patch to the Axes
#ax.add_patch(patches.Rectangle((0,0),gap/2,gap/2,linewidth=10,edgecolor='r',facecolor='none'))


ax.set_aspect('equal')
plt.xlim((-1, gap*nx + gap))
plt.ylim((-1, gap*ny + gap))
plt.axis("off")
```

```python
def plot_arr(arr, ax, gap, fill_x_start=None, fill_x_end=None, fill_y_start=None, fill_y_end=None, fill_color='purple',bg_color='pink'):
    nx, ny = arr.shape
    if fill_x_end is None or fill_y_end is None or fill_x_start is None or fill_y_start is None:
        fill_x_start = 0
        fill_x_end = arr.shape[0]
        fill_y_end =arr.shape[1]
        fill_y_start = 0 
    rects = np.zeros((nx, ny)).tolist()
    for i in range(nx):
        for j in range(ny):
            if fill_x_start<=i<=fill_x_end and fill_y_start<=j<=fill_y_end:
                rects[i][j] = patches.Rectangle((i*gap, j*gap),gap,gap,linewidth=1,edgecolor='k',facecolor=fill_color,alpha=0.5)
            else:
                rects[i][j] = patches.Rectangle((i*gap, j*gap),gap,gap,linewidth=1,edgecolor='k',facecolor=bg_color, alpha=0.3)
            ax.add_patch(rects[i][j])
            ax.text(i*gap+gap/3, j*gap+gap/3, arr[i, j])

    ax.set_aspect('equal')
    ax.set_xlim((-1, gap*nx + gap))
    ax.set_ylim((-1, gap*ny + gap))
    ax.axis("off")
    ax.invert_yaxis()
```

```python
fig, ax = plt.subplots()
plot_arr(np.round(np.random.randn(8, 6)), ax, 4, bg_color='green', fill_color='pink')
```

```python
fig, ax = plt.subplots()
plot_arr(np.round(np.random.randn(8, 6)), ax, 4,  
         fill_x_start=1, fill_x_end=4, fill_y_start=2, fill_y_end=4)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(10, 6))
np.random.seed(0)
input_arr = np.round(np.random.randn(6, 6))
filter_arr =np.round(np.random.randn(3, 3))
stride = 1
out_arr = np.zeros((4, 4))

nx_inp, ny_inp = input_arr.shape
nx_fil, ny_fil = filter_arr.shape
for i in range(nx_inp-nx_fil):
    for j in range(ny_inp-ny_fil):
        out_arr[i, j] = np.multiply(input_arr[i:i+nx_fil, j:j+ny_fil], filter_arr).sum()
        plot_arr(input_arr, ax[0], 10, fill_x_end=i+nx_fil, fill_x_start=i, fill_y_start=j, fill_y_end=j+ny_fil)
        plot_arr(filter_arr, ax[1], 2)
        plot_arr(out_arr, ax[2], 2)


plt.tight_layout()
```
