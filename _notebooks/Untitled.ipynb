```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.reset_defaults()
sns.set_context(context='talk',font_scale=1)
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
s = np.array([0, 6, 24, 36])
t = np.array([0, 1, 3, 4])
t_sqr = t**2
```

```python
plt.plot(t, s, marker='*')
plt.xlabel("t")
plt.ylabel("s")
sns.despine()
plt.title("s v/s t")
```

```python
plt.plot(t_sqr, s, marker='*')
plt.xlabel("t")
plt.ylabel("s")
sns.despine()
plt.title("s v/s t^2")
```
